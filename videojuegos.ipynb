Importamos las librerías que vamos a utilizar

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud,  STOPWORDS
from PIL import Image
import numpy as np 
import PIL.Image
```

Utilizamos pandas para hacer Web scraping al sitio web de wikipedia y extraer datos de la tabla que deseamos

```python
tabla=pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_video_games")
resultado=tabla[1]
tablafinal=resultado.drop(columns=["Ref."])
tablafinal
```

Con la librería Matplotlib podemos generar gráficos para visualizar datos interesantes

```python
top_10=tablafinal.sort_values(by='Sales', ascending=False).head(10)

title=top_10['Title']
sales=top_10['Sales']


plt.pie(sales, labels=title )
plt.title("TOP 10 VIDEOJUEGOS MAS VENDIDOS")
plt.show()
```

Realizamos un Wordcloud de los desarroladores

```python
desarolladores=tablafinal['Developer(s)[b]']
desarrolladores_texto = ' '.join(desarolladores.dropna())


wc=WordCloud(stopwords=STOPWORDS, background_color='white').generate(desarrolladores_texto)

plt.imshow(wc)
plt.axis('off')
plt.title("WORCLOUD DEVELOPERS")
plt.show()
```

Terminamos con un worcloud en forma de mando para hacerlo más atractivo

```python
juegos=tablafinal['Title']
juegos_texto = ' '.join(juegos.dropna())
mask = np.array(Image.open('mando.jpg'))

wc1=WordCloud(stopwords=STOPWORDS, background_color='black', mask=mask).generate(juegos_texto)
plt.imshow(wc1)
plt.axis('off')
plt.title("WORCLOUD  VIDEOGAMES")
plt.show()
```
